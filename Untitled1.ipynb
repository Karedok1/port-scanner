```python
import socket
import threading
from datetime import datetime



print("="*33)
print("    SIMPLE PORT SCANNER")
print("="*33)

ip_input = input(("masukan alamat ip (pisahkan dengan koma): "))
ip_list = [ip.strip() for ip in ip_input.split(",")]
print(f"ip list {ip_list}")





awal = int(input("masukan port: "))
akhir = int(input("masukan port: "))
thread_list = []
pertama = datetime.now()


def scan_port(ip, port):
    try:
        service = socket.getservbyport(port)

    except Exception:
        service = "unknown"
        pass



    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.settimeout(0.7)
    result = s.connect_ex((ip, port))
    if result == 0:
        print(f"{ip} port {port} terbuka {service.upper()}")

    else:
        print(f"{ip} tertutup {port}")
    s.close()


for ip in ip_list:
    for port in range(awal, akhir + 1):
        thread = threading.Thread(target=scan_port, args=(ip, port))
        thread.start()
        thread_list.append(thread)

for t in thread_list:
    t.join()


selesai = datetime.now()
i = selesai - pertama
print(i)
```
